```python
#리스트 메서드 직접 구현

#(1) len()
def get_len(input_list) :
    cnt = 0
    for _ in input_list :
        cnt += 1
    return cnt
print(get_len([1,2,3,3])) #4
 
#(2) count()
def get_count(input_list, find_value) :
    count = 0
    for item in input_list :
        if item == find_value:
            count += 1
    return count

print(get_count([1,2,3,3], 3)) #2


#리스트 관련 함수 직접 구현
#(2) min(), max()
def get_min_max(input_list) :
    min_value = input_list[0]
    max_value = input_list[0]
    
    for item in input_list :
        if item < min_value:
            min_value = item
        elif item > max_value:
            max_value = item
            
    return min_value, max_value
print(get_min_max([2,3,4,11,0])) #(0,11)

#(3) sum()
def get_sum(input_list) :
    total = 0
    for item in input_list :
        total += item
    return total
    
#(3) avg() : 평균 구하는 함수
def get_avg(input_list) :
    total = sum(input_list)
    avg = total/len(input_list)
    
    return float(avg)

print(get_avg([1,2,3,4,5]))

#(4) std() : 표준편차 구하는 함수 (편차 제곱의 평균)
def get_std(input_list) :
    avg = sum(input_list) / len(input_list)
    variance = sum((i - avg) ** 2 for i in input_list) / (len(input_list) - 1)
    std_dev = variance ** 0.5
    return std_dev
    
 

print(get_std([1,2,3,4,5]))

#(5) median() : 중앙값 구하는 함수
#오름차순 정렬 후 
#데이터의 개수가 홀수일 때 : 인덱스(데이터의 개수 // 2)의 값
#ex) 1,2,3,4,5 => 3
#데이터의 개수가 짝수일 때 : (인덱스(데이터의 개수 // 2 -1)의 값 + 인덱스(데이터의 개수 // 2 )의 값) / 2
#ex) 1,2,3,4 => (2+3)/2
def get_median(input_list) :
    sort_list = sorted(input_list)
    
    leng = len(sort_list)
    middle = leng//2
    
    if leng % 2 == 0 :
        median = (sort_list[middle - 1]+ sort_list[middle]) / 2
    else :
        median = sort_list[middle]
        
    return median
print(get_median([1,2,3,4]))
print(get_median([1,2,3,4,5]))
```

```python
#문자열 조작 메서드 직접 구현
#(1-1) find() : find_value는 문자 한개를 입력으로 일치하는 인덱스 반환, 없으면 -1 반환  
def get_find_ch(text, find_value) :
    for i, j in enumerate(text):
        if j == find_value:
            return i
    return -1
print(get_find_ch('hello', 'l')) #2

#(1-2) find() : find_value는 문자열 입력으로 일치하는 인덱스 반환, 없으면 -1 반환  
def get_find_str(text, find_str, start=0) :
    for index in range(start, len(text) - len(find_str) + 1):
        if text[index:index+len(find_str)] == find_str:
            return index
    return -1
print(get_find_str('hello', 'llo')) #2

#(2-1) split() : sep는 문자 한개를 입력으로 구분자대로 잘라서 리스트로 반환
def str_split(text, sep=' ') :
    result = []
    start = 0
    
    for i, j in enumerate(text):
        if j == sep:
            result.append(text[start:i])
            start = i+1
            
    result.append(text[start:])
    return result

print(str_split('hello my name is gildong',  ' '))  #['hello', 'my' ,'name', 'is', 'gildong']

#(2-2) split() : sep는 문자열을 입력으로 구분자대로 잘라서 리스트로 반환
def str_split(text, sep='--') : 
    result = []
    start = 0
    
    while True :
        index = text.find(sep, start)
        if index == -1:
            break
        
        result.append(text[start:index])
        start = index+len(sep)
        
    result.append(text[start:])
    
    return result
    
print(str_split('hello my name is  gildong ',  '  ')) #['hello my name is', 'gildong ']

#(3) replace() : text에서 old문자열을 찾아서 new로 변경
def str_replace(text, old, new) :
    result = ''
    i = 0
    
    while i < len(text):
        if text[i:i+len(old)] == old:
            result += new
            i += len(old)
        else:
            result += text[i]
            i += 1
    return result

print(str_replace('hello my name is gildong', 'gildong', 'hani'))  #'hello my name is hani'

#(4) strip() : text에서 앞뒤로 remove_ch와 일치하는 문자를 제거
# remove_ch는 문자 한개를 입력으로 함
def str_strip(text, remove_ch) : 
    start = 0
    end = len(text)
    
    while start < end and text[start] == remove_ch:
        start +=1
        
    while end > start and text[end-1] == remove_ch:
        end -=1
    return text[start:end]
print(str_strip('----hello--', '-')) #hello

#첫글자가 영어의 소문자일 경우 대문자로 변경
#인터넷에 아스키 코드 참조하여 이용
#ord(문자한개) : 아스키코드값으로 반환
#chr(아스키코드) : 해당 아스키코드값을 문자로 반환
#(5) title()
def str_title(text) :
    first_char_ascii = ord(text[0])
    
    if 'a' <= text[0] <= 'z':
        first_char_ascii -= 32
    return chr(first_char_ascii)+ text[1:]
    
print(str_title('hello')) #Hello
```
